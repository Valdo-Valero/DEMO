```python
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder.getOrCreate()
```

```python
df1 = spark.read.csv("TSLA-1.csv", inferSchema= True, header = True)
df1.show(truncate = False)

df2 = spark.read.csv("TSLA-2.csv", inferSchema= True, header = True)
df2.show(truncate = False)
```

```python
import pyspark.sql.functions as F
split_col = F.split(df1['Date,Open,High,Low,Close,Adj�Close,Volume'],',')
df1 = df1.withColumn('Date', split_col.getItem(0))\
.withColumn('Open', split_col.getItem(1))\
.withColumn('High', split_col.getItem(2))\
.withColumn('Close', split_col.getItem(3))\
.withColumn('Adj Close', split_col.getItem(5))\
.withColumn('Volume', split_col.getItem(6))\

df1.show(50)

split_col = F.split(df2['Date,Open,High,Low,Close,Adj�Close,Volume'],',')
df2 = df2.withColumn('Date', split_col.getItem(0))\
.withColumn('Open', split_col.getItem(1))\
.withColumn('High', split_col.getItem(2))\
.withColumn('Close', split_col.getItem(3))\
.withColumn('Adj Close', split_col.getItem(5))\
.withColumn('Volume', split_col.getItem(6))\

df2.show(50)
```

```python
df3 = df1.union(df2)
df3.show(50)
```

```python
sele_column = ["Date", "Open", "High", "Close", "Adj Close", "Volume"]
df3 = df3.select(sele_column)
df3.show(50)
```

```python
df3 = df3.where(df3.Volume > 50000000)
```

```python
df3.sort('Date').show(truncate = False)
```
